# Log 변환
* NaN 값이 나와서 사용하지 않기로 함

```python
# 라이브러리 로드
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 데이터 로딩
df = pd.read_csv('./data/2011_2020_m2.ver2.csv', encoding='utf-8')
df
```

```python
# 데이터셋 구성
df_tgt = df[['WIND_SPEED', 'GUST_WIND', 'SIG_WAVE', 'MAX_WAVE', 'HPA']]
```

* 로그

```python
# 로그 
log_df_tgt=np.log(df_tgt)
```

```python
log_df_tgt
```

```python
# 다시 '일시', '특보강도', '남은시간' 합침
df1 = pd.concat([log_df_tgt,df[['일시','특보강도','남은시간']]], axis=1)
df1
```

```python
print('feature 들의 평균 값')
print(df1.mean())
print('\nfeature 들의 분산 값')
print(df1.var())
```
