```python
# 라이브러리 로드
import pandas as pd
import numpy as np
```

```python
# 데이터 로딩
df = pd.read_csv('./data/2011_2020_m2.ver2.csv', encoding='utf-8')
df
```

```python
print(df["특보강도"].value_counts())
```

```python
# obj -> datetime
df['일시'] = pd.to_datetime(df['일시'])
df.info()
```

```python
# 인덱스
df = df.set_index('일시')
df
```

```python
# 데이터셋 구성
df_tgt = df[['WIND_SPEED', 'GUST_WIND', 'SIG_WAVE', 'MAX_WAVE', 'HPA']]
```

* StandardScaler

```python
# standardscaler 라이브러리 로드
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df_tgt)
```

```python
arr = scaler.transform(df_tgt)
```

```python
df1 = pd.DataFrame(arr, columns=df_tgt.columns)
df1
```
