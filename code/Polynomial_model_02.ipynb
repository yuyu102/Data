```python
# PolynomialFeatures 라이브러리 호출
from sklearn.preprocessing import PolynomialFeatures

import pandas as pd
```

```python
df = pd.read_csv('./StandardScaler.csv', encoding='utf-8')
df
```

```python
df_tgt = df[['WIND_SPEED', 'GUST_WIND', 'SIG_WAVE', 'MAX_WAVE', 'HPA']]
```

```python
poly_features = PolynomialFeatures(degree=2)
```

```python
df_poly = poly_features.fit_transform(df_tgt)
```

```python
df_poly = pd.DataFrame(df_poly)
```

```python
df_poly #'1','x1','x2','x1^2','x1*x2','x2^2'
```

```python
df1 = pd.concat([df_poly,df[['일시','특보강도']]], axis=1)
df1
```

```python
df1['일시'] = pd.to_datetime(df1['일시'])
df1.info()
```

```python
df1 = df1.set_index('일시')
df1
```

```python
df1.to_csv('Standard_Poly.csv')
```
