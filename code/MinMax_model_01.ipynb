```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/2011_2020_m2.ver2.csv', encoding='utf-8')
df
```

```python
df_tgt = df[['WIND_SPEED', 'GUST_WIND', 'SIG_WAVE', 'MAX_WAVE', 'HPA']]
```

```python
print('feature들의 최소 값')
print(df_tgt.min())
print('\nfeature들의 최대 값')
print(df_tgt.max())
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# MinMaxScaler객체 생성
scaler = MinMaxScaler()
```

```python
# # MinMaxScaler 로 데이터 셋 변환. fit() 과 transform() 호출.  
# scaler.fit(df_tgt)
# df_scaled = scaler.transform(df_tgt)
```

```python
scaler.fit(df_tgt)
```

```python
arr = scaler.transform(df_tgt)
```

```python
df1 = pd.DataFrame(arr, columns=df_tgt.columns)
df1
```

```python
df1 = pd.concat([df1,df[['일시','특보강도','남은시간']]], axis=1)
df1
```

```python
print('feature 들의 평균 값')
print(df1.mean())
print('\nfeature 들의 분산 값')
print(df1.var())
```
